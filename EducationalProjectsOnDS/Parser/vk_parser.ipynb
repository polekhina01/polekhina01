### Пишем функцию получения данных о подписчиках группы ВК

```python
import requests

# объявляем переменные 
TOKEN_USER = 'ВСТАВЬТЕ СЮДА СВОЙ ТОКЕН'
SORT = 'id_asc'
VERSION = '5.154'

# через api vk собираем статистику подписчиков группы
def group_getMembers(TOKEN_USER, GROUP_ID, SORT, OFFSET, COUNT, FIELDS, FILTER, VERSION):
    response = requests.get('https://api.vk.com/method/groups.getMembers',
    params={'access_token': TOKEN_USER,
            'group_id': GROUP_ID,
            'sort': SORT,
            'offset': OFFSET,
            'count': COUNT,
            'fields': FIELDS,
            'filter': FILTER,
            'v': VERSION})

    data = response.json()['response']['items']
    return data
```

### Вызываем функцию, смотрим что получилось

```python
merged_data = []

offset = 0
for i in range(3):
    data = group_getMembers(TOKEN_USER = TOKEN_USER, 
                            GROUP_ID = '203350625', 
                            SORT = SORT, 
                            OFFSET = offset, 
                            COUNT = 1000, 
                            FIELDS = 'contacts, connections, site, education, universities, schools',
                            FILTER = '',
                            VERSION = VERSION)
    offset += 1000
    merged_data = merged_data + data
```

### Создаем датафрейм и записываем данные в xlsx файл

```python
import pandas as pd

df = pd.DataFrame(merged_data)

df.to_excel('dataframe.xlsx', index=False)
```

### Пишем функцию для сбора id-шников друзей каждого подписчика

```python
def get_friends(user_id, access_token):
    url = f"https://api.vk.com/method/friends.get?user_id={user_id}&access_token={access_token}&v=5.131"
    response = requests.get(url)
    data = response.json()
    
    if 'response' in data:
        friend_ids = data['response']['items']
        return friend_ids
    else:
        return []
    

friends = get_friends(user_id = '222476348', access_token=TOKEN_USER)
```

### Пишем результаты в json файл

```python
import json

# Создаем словарь для хранения списков друзей
friends_dict = {}

# Для каждого элемента колонки id получаем список друзей и добавляем в словарь
for user_id in df['id']:
    friends = get_friends(user_id, access_token=TOKEN_USER)
    friends_dict[user_id] = friends

# Записываем словарь в файл в формате JSON
with open('friends.json', 'w') as f:
    json.dump(friends_dict, f)
```
